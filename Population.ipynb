```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a sample dataset
np.random.seed(0)
n_samples = 1000
```

```python
# Generate sample data
ages = np.random.normal(35, 15, n_samples).astype(int)
ages = np.clip(ages, 0, 100)
genders = np.random.choice(['Male', 'Female', 'Other'], n_samples, p=[0.49, 0.49, 0.02])
df = pd.DataFrame({
    'Age': ages,
    'Gender': genders
})

# Figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Bar chart for gender distribution
gender_counts = df['Gender'].value_counts()
ax1.bar(gender_counts.index, gender_counts.values)
ax1.set_title('Distribution of Genders')
ax1.set_xlabel('Gender')
ax1.set_ylabel('Count')

# Add value labels on top of each bar
for i, v in enumerate(gender_counts.values):
    ax1.text(i, v, str(v), ha='center', va='bottom')

# Histogram for age distribution
ax2.hist(df['Age'], bins=20, edgecolor='black')
ax2.set_title('Distribution of Ages')
ax2.set_xlabel('Age')
ax2.set_ylabel('Count')

# Adjust layout and display the charts
plt.tight_layout()
plt.show()
```
